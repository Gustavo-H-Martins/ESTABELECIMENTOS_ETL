```python
import requests

url = 'https://geoftp.ibge.gov.br/organizacao_do_territorio/estrutura_territorial/localidades/Geomedia_MDB/BR_Localidades_2010_v1.mdb' 
r = requests.get(url)
part = url.split('/')
with open(part[-1], 'wb') as f:
    f.write(r.content)
```

```python
import pandas as pd
caminho = './BR_Localidades_2010_v1.xlsx'

dados = pd.read_excel(caminho)
```

```python
localidades = dados[['NM_LOCALIDADE', 'NM_UF', 'LONG', 'LAT']]
localidades.rename(columns={'NM_LOCALIDADE':'DISTRITO', 'NM_UF':'ESTADO', 'LONG':'LONGITUDE', 'LAT':'LATITUDE'}, inplace=True)
localidades
```

```python
localidades.to_csv(r'./localidades.txt', index=False, sep='\t', mode='a')
```
