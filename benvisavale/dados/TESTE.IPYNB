```python
import pandas as pd
dados = pd.read_csv('./BASE_BENVISAVALE.csv', sep=';')
dados.head(5)
```

```python
dados.rename(columns={'BENVISAVALE': 'BANDEIRA'}, inplace=True)
```

```python
LONGITUDE = []
for i in dados['LONGITUDE BANDEIRA']:
    dado = i.split(' ')
    LONGITUDE.append(dado[0])
dados['LONGITUDE'] = LONGITUDE
```

```python
dados.head(2)
```

```python
BAIRRO = []
MUNICIPI = []

for i in dados['BAIRRO, UF, MUNICIPIO, CEP-TELEFONE']:
    I = i.split(', ')
    BAIRRO.append(I[0])
    MUNICIPI.append(I[2])
```

```python
dados['BAIRRO'] = BAIRRO
dados['MUNICIPIO'] = MUNICIPI
```

```python
dados = dados[['ESTABELECIMENTOS','ENDERECO','BAIRRO','MUNICIPIO','UF','CEP','TELEFONE','EMAIL','LATITUDE','LONGITUDE','BANDEIRA', 'CIDADE_PADRAO_IBGE']]
```

```python
dados.to_csv('./BASE_BENVISAVALE.csv', sep=';', index=False)
```

```python
dados.to_excel('./BASE_BENVISAVALE.xlsx', sheet_name='BASE BEN VISA VALE', index=False)
```
